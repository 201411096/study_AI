### 정보의 종류에 따른 시각화 유형

* 히스토그램
    * 연속형 값에 대한 도수 분포를 나타냄. X축값은 도수 분포를 원하는 연속형 값의 구간, Y축값은 해당 구간의 도수 분포
* 바이올린 플롯
    * 어떤 값의 분포를 특정 (이산)값 별로 시각화 할 때 유용하게 활용됨
* 바 플롯
    * X축값은 이산값, Y축값은 연속형값으로 특정 이산값 X별로 연속형 Y값을 보고자 할 때 사용
* 스캐터 플롯
    * X축값은 연속형 값, Y축값도 연속형값일 때 해당 값들이 어떻게 분포되어 있는지 시각화 할 때 유용
* 라인 플롯
    * 시계열 데이터를 시각화하는데 유용. 보통 X축에 년/월/일, 그리고 Y축에 연속형 값으로 시각화
